```python
data_no2 = data.dropna(subset=['no2'])
data_so2 = data.dropna(subset=['so2'])
plt.figure(figsize=(14,6))
plt.scatter(list(range(data_no2.loc[data_no2['location']=='Delhi', 'no2'].values.shape[0])) , data_no2.loc[data_no2['location']=='Delhi', 'no2'].values, label='NO2, Delhi')
plt.plot(data_no2.loc[data_no2['location']=='Delhi', 'no2'].rolling(100).mean().values, linewidth=4, c='r', label='NO2 100 avg days, Delhi')
plt.scatter(list(range(data_so2.loc[data_so2['location']=='Delhi', 'so2'].values.shape[0])), data_so2.loc[data_so2['location']=='Delhi', 'so2'].values, c='y', label='SO2, Delhi')
plt.plot(data_so2.loc[data_so2['location']=='Delhi', 'so2'].rolling(50).mean().values, linewidth=4, c='g', label='SO2 100 avg days, Delhi')
plt.legend(loc='upper left')
plt.ylabel('ppm')
plt.xlabel('days')
plt.show()
```

```python
data_no2 = data.dropna(subset=['no2'])
data_so2 = data.dropna(subset=['so2'])
plt.figure(figsize=(14,6))
plt.plot(data_no2.loc[data_no2['type']=='Industrial Area', 'no2'].rolling(500).mean().values, label='NO2, Industrial Area')
plt.plot(data_no2.loc[data_no2['type']=='Residential, Rural and other Areas', 'no2'].rolling(500).mean().values,
                                                                 linewidth=4, label='NO2, Residential, Rural and other Areas')
plt.plot(data_so2.loc[data_so2['type']=='Industrial Area', 'so2'].rolling(500).mean().values, label='SO2, Industrial Area')
plt.plot(data_so2.loc[data_so2['type']=='Residential, Rural and other Areas', 'so2'].rolling(500).mean().values, 
                                                                 linewidth=4, label='SO2, Residential, Rural and other Areas')
plt.legend(loc = 'upper right')
plt.show()
```
